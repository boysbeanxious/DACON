```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('./train.csv')
```

```python
train.head()
```

```python
train.dtypes
```

```python
train['송하인_격자공간고유번호'] = train['송하인_격자공간고유번호'].astype('str')
train['수하인_격자공간고유번호'] = train['수하인_격자공간고유번호'].astype('str')
```

```python
train['송하인_격자공간고유번호_sigungu'] = train['송하인_격자공간고유번호'].str.slice(0, 5)
```

```python
train['수하인_격자공간고유번호_sigungu'] = train['수하인_격자공간고유번호'].str.slice(0, 5)
```

```python
train.head()
```

```python
sigungu = pd.read_csv('./국토교통부_전국 법정동_20211217_필터.csv', encoding='cp949')
```

```python
sigungu.shape
```

```python
sigungu.dtypes
```

```python
sigungu['법정동코드'] = sigungu['법정동코드'].astype('str')
```

```python
sigungu['sigungu'] = sigungu['법정동코드'].str.slice(0, 5)
                                 
```

```python
sigungu = sigungu[[ '시도명', '시군구명', 'sigungu']].drop_duplicates().reset_index(drop = True)
```

```python
train_merge = pd.merge(train, sigungu, left_on = ['송하인_격자공간고유번호_sigungu'], right_on = ['sigungu'], how = "left")
```

```python
print(train.shape)
print(train_merge.shape)
```

```python
train_merge.head()
```

```python
train_merge = train_merge[['index', '송하인_격자공간고유번호', '수하인_격자공간고유번호', '물품_카테고리', '운송장_건수',
       '송하인_격자공간고유번호_sigungu', '수하인_격자공간고유번호_sigungu', '시도명', '시군구명']]
```

```python
train_merge.columns = ['index', '송하인_격자공간고유번호', '수하인_격자공간고유번호', '물품_카테고리', '운송장_건수',
       '송하인_격자공간고유번호_sigungu', '수하인_격자공간고유번호_sigungu', '송하인_시도명', '송하인_시군구명',]
```

```python
train_merge
```

```python
train_df = pd.merge(train_merge, sigungu, left_on = ['수하인_격자공간고유번호_sigungu'], right_on = ['sigungu'], how = "left")
```

```python
train_df = train_df[['index', '송하인_격자공간고유번호', '수하인_격자공간고유번호', '물품_카테고리', '운송장_건수',
       '송하인_격자공간고유번호_sigungu', '수하인_격자공간고유번호_sigungu', '송하인_시도명', '송하인_시군구명',
       '시도명', '시군구명']]
```

```python
train_df.columns = ['index', '송하인_격자공간고유번호', '수하인_격자공간고유번호', '물품_카테고리', '운송장_건수',
       '송하인_격자공간고유번호_sigungu', '수하인_격자공간고유번호_sigungu', '송하인_시도명', '송하인_시군구명',
       '수하인_시도명', '수하인_시군구명']
```
