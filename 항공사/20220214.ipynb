```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('./train.csv')
test = pd.read_csv('./test.csv')
submission = pd.read_csv('./sample_submission.csv')
```

```python
train.head()
```

```python
pd.get_dummies()
```

```python
obj_arr = np.array(train.dtypes[pd.DataFrame(train.dtypes)[0]== 'object'].index)
```

```python
train_dummy = pd.get_dummies(train, columns = obj_arr)
```

```python
train_corr = train_dummy[['Age', 'Flight Distance', 'Seat comfort',
       'Departure/Arrival time convenient', 'Food and drink', 'Gate location',
       'Inflight wifi service', 'Inflight entertainment', 'Online support',
       'Ease of Online booking', 'On-board service', 'Leg room service',
       'Baggage handling', 'Checkin service', 'Cleanliness', 'Online boarding',
       'Departure Delay in Minutes', 'Arrival Delay in Minutes', 'target',
       'Gender_Female', 'Gender_Male', 'Customer Type_Loyal Customer',
       'Customer Type_disloyal Customer', 'Type of Travel_Business travel',
       'Type of Travel_Personal Travel', 'Class_Business', 'Class_Eco',
       'Class_Eco Plus']].corr().reset_index()
```

```python
train_corr
```

```python
train_corr = train_corr.rename(columns={'index':'columns'})
```

```python
train_corr[train_corr['columns']=='target']
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
X_train = pd.DataFrame(train['Arrival Delay in Minutes'])
y_train =train['target']
```

```python
X_test = pd.DataFrame(test['Arrival Delay in Minutes'])
```

```python
model = lr.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
submission['target'] = y_pred
```

```python
submission.to_csv("submission.csv",index=False)
```
